```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("Ministry of Fisheries and Livestock Zambia.xlsx")
```

```python
df
```

#### Cleaning the Data

```python
df.columns
```

```python
 df.rename(columns={"MONTH\n": "DATE","STEAK/KG":"STEAK","MIXED\nCUT/KG":"MIXED-CUT","BEEF\nOFFALS/KG":"BEEF-OFFALS",'HOOVES/KG':'HOOVES','PORK/KG':'PORK','T-BONE/KG':'T-BONE',"GOAT/KG":"GOAT"},inplace=True)
```

```python
df
```

```python
df["Number"] = range(1,11)
```

```python
df.set_index("Number",inplace = True)
```

#### create a year and month column

```python
df['DATE'] = df['DATE'].astype(str)
```

```python
df["YEAR"] = df["DATE"].str[0:4]
```

```python
df["MONTH"] = df["DATE"].str[5:7]
```

```python
df["DAY"] = df["DATE"].str[8:10]
```

```python
df
```

```python
df.drop("COMMENTS",axis=1,inplace=True)
```

```python
df
```

#### 1) What was the highest price for a particular product, and on which date was it recorded?

```python
df.columns
```

```python
prompt = input("Enter the name of the product: ").upper()
df_max = df[prompt].max()
max_loc = df.loc[(df[prompt] == df[prompt].max())][["YEAR","MONTH","DAY"]]
print(f"The highest price for {prompt} was {df_max}")
print(f"recorded on \n{max_loc}")
```

```python
bars = df.groupby(["MONTH"]).sum()
print(bars)
```

```python
months = ["January","February","March","April","July","August","September","Ocotber","November","December"]
prompt = input("Enter the name of the product: ").upper()
plt.bar(months, bars[prompt], color = "royalblue")
plt.title(f"{prompt} (kg)")
plt.xlabel("Months")
plt.ylabel("prices in Zambian Kwacha")
plt.xticks(months,rotation = "vertical")
#plt.legend()
plt.show()
```

#### 4) What was the percentage change in price for a product between the months of July and April?

```python
df
```

###### percentage_change = ((new_value - old_value) / abs(old_value)) * 100

##### Steak

```python
prompt = input("Enter the name of the meat product: ").upper()
percentage_change = (df.loc[10][prompt] - df.loc[1][prompt]) / abs(df.loc[1][prompt]) * 100
print(f"The percentage increase was {percentage_change} percent.")
```

##### Hooves  

```python
prompt = input("Enter the name of the meat product: ").upper()
percentage_change = (df.loc[10][prompt] - df.loc[1][prompt]) / abs(df.loc[1][prompt]) * 100
print(f"The percentage increase was {percentage_change} percent.")
```

#### 5) Among these products, which one exhibited the most consistent pricing?

```python
df.columns
```

```python
consistent_pricing=[df['STEAK'].max() -df['STEAK'].min() , df['MIXED-CUT'].max() - df['MIXED-CUT'].min(), df['T-BONE'].max() - df['T-BONE'].min(),df['BEEF-OFFALS'].max()-df['BEEF-OFFALS'].min(), df['GOAT'].max()-df['GOAT'].min(), df['PORK'].max()-df['PORK'].min(),
       df['HOOVES'].max()-df['HOOVES'].min()]
print(f"Mixed-Cut has a price range of {min(consistent_pricing)} hence the most consistent pricing.")
products = ['STEAK', 'MIXED-CUT', 'T-BONE', 'BEEF-OFFALS', 'GOAT', 'PORK',
       'HOOVES']
plt.bar(products,consistent_pricing)
plt.xticks(products,rotation = "vertical")
plt.title("Price Range")
plt.show()
```
